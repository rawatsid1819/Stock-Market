```python
hdfc["Return"]=hdfc["Close"].pct_change(1)
```

```python
hdfc["Return"].hist(label="hdfc",bins=50)
sbi["Return"].hist(label="sbi",bins=50)
icici["Return"].hist(label="icici",bins=50)
plt.legend()
```

```python
import matplotlib.pyplot as plt
fig = plt.gcf()
fig.set_size_inches(12, 8)
sns.distplot(hdfc["Return"],label="hdfc",kde=False)
sns.distplot(sbi["Return"],label="sbi",kde=False)
sns.distplot(icici["Return"],label="icici",kde=False)
plt.legend()
```

```python
sns.distplot(hdfc["Return"],label="hdfc",kde=False)
sns.distplot(sbi["Return"],label="sbi",kde=False)
sns.distplot(icici["Return"],label="icici",kde=False)
plt.legend()
```

```python
hdfc["Compund Return"]=(hdfc["Return"]+1).cumprod()
```

```python
sbi["Compund Return"]=(sbi["Return"]+1).cumprod()
icici["Compund Return"]=(icici["Return"]+1).cumprod()
```

```python
hdfc
```

```python
hdfc["Compund Return"].plot(figsize=(16,8),label="hdfc",title="Compound Return")
sbi["Compund Return"].plot(label="sbi")
icici["Compund Return"].plot(label="icici")
plt.xlabel("time index")
plt.legend()
```
